### Happy Number
Write an algorithm to determine if a number **n** is happy.

A **happy number** is a number defined by the following process:

* Starting with any positive integer, replace the number by the sum of the squares of its digits.
* Repeat the process until the number equals 1 (where it will stay), or it **loops endlessly in a cycle** which does not include 1.
* Those numbers for which this process **ends in 1** are happy.

Return **true** if n is a happy number, and **false** if not.

Constraints:
1 <= n <= $2^{31}$ - 1

```python
def isHappy(n):
  # 필요한 변수 및 데이터들 선언 및 전처리 과정
  before = [n]                                # 무한 반복인지 확인하기 위한 리스트; 이전에 있었던 수들을 저장
  n = list(map(int, list(str(n))))            # 한 자리씩 떼어서 리스트에 저장하기


  # 조건이 만족할 때까지 반복
  while True:
    # 한 자리씩 제곱해서 더하는 계산 부분
    temp = 0
    for i in range(len(n)):
      temp += n[i] ** 2
    n = list(map(int, list(str(temp))))

    # 멈춤 조건 판별
    if temp == 1:                            # 멈추는 조건 1; 1
      return True
      break
    elif temp in before:                     # 멈추는 조건 2: 과거에 있었던 수
      return False
      break
    else:
      before.append(temp)


n = int(input())                              # input의 형태: 숫자만 하나 들어옴
print(isHappy(n))
```
